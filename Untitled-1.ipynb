```python
import pandas as pd
import matplotlib.pyplot as plt

def analisar_fontes_energia(caminho_arquivo, fontes_energia, titulo_grafico, coluna_ano):
    """
    Carrega dados de um arquivo CSV, filtra as colunas relevantes e plota um gr√°fico de linhas.
    """
    def carregar_dados(caminho_arquivo):
        """
        Carrega o arquivo CSV e remove espa√ßos extras nos nomes das colunas.
        """
        try:
            dados = pd.read_csv(caminho_arquivo)
            dados.columns = dados.columns.str.strip()  
            return dados
        except Exception as e:
            print(f"Erro ao carregar os dados: {e}")
            return None

    def filtrar_dados(dados, fontes_energia):
        """
        Filtra as colunas de fontes de energia de interesse.
        """
        return dados[fontes_energia]

    def gerar_grafico(dados_filtrados, fontes_energia, titulo_grafico, coluna_ano):
        """
        Gera um gr√°fico de linhas para a evolu√ß√£o das fontes de energia.
        """

        dados_filtrados.plot(kind='line', marker='o', figsize=(10, 6))
        plt.title(titulo_grafico)
        plt.xlabel('Ano')  
        plt.ylabel('Participa√ß√£o (%)')
        plt.xticks(ticks=range(len(dados_filtrados)), labels=dados[coluna_ano], rotation=45)
        plt.legend(title='Fontes de Energia', labels=fontes_energia)
        plt.tight_layout()
        plt.show()

    dados = carregar_dados(caminho_arquivo)
    if dados is not None:
        dados_filtrados = filtrar_dados(dados, fontes_energia)
        gerar_grafico(dados_filtrados, fontes_energia, titulo_grafico, coluna_ano)

caminho_arquivo_1 = "1.csv"  
caminho_arquivo_2 = "3.csv"  

fontes_energia_1 = ['Coal - % electricity', 'Gas - % electricity', 'Hydro - % electricity', 'Solar - % electricity', 'Wind - % electricity', 'Bioenergy - % electricity']
fontes_energia_2 = ['Nuclear - % electricity', 'Fossil fuels - % electricity', 'Renewables - % electricity']

coluna_ano_1 = 'Year'  
coluna_ano_2 = 'Year'  

print("Gr√°fico para o primeiro dataset:")
analisar_fontes_energia(caminho_arquivo_1, fontes_energia_1, 'Evolu√ß√£o das Fontes de Energia - Dataset 1', coluna_ano_1)

print("Gr√°fico para o segundo dataset:")
analisar_fontes_energia(caminho_arquivo_2, fontes_energia_2, 'Evolu√ß√£o das Fontes de Energia - Dataset 2', coluna_ano_2)
```
