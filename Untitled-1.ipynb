```python
import pandas as pd
import matplotlib.pyplot as plt

def carregar_dados(caminho_arquivo):
    dados = pd.read_csv(caminho_arquivo)
    dados.columns = dados.columns.str.strip()  
    return dados

def filtrar_participacao(dados, fontes_energia):
    dados_filtrados = dados[fontes_energia]
    return dados_filtrados

def gerar_grafico_comparacao(dados_filtrados, fontes_energia):
    dados_filtrados.plot(kind='bar', stacked=True, figsize=(10, 6))
    plt.title('Compara√ß√£o entre Fontes de Energia')
    plt.xlabel('Pa√≠s')
    plt.ylabel('Participa√ß√£o (%)')
    plt.xticks(rotation=90)
    plt.legend(title='Fontes de Energia', labels=fontes_energia)
    plt.tight_layout()
    plt.show()

caminho_arquivo_1 = "1.csv" 
caminho_arquivo_2 = "3.csv"  

dados_1 = carregar_dados(caminho_arquivo_1)
dados_2 = carregar_dados(caminho_arquivo_2)

fontes_energia_1 = ['Coal - % electricity', 'Gas - % electricity', 'Hydro - % electricity', 'Solar - % electricity', 'Wind - % electricity', 'Bioenergy - % electricity']

fontes_energia_2 = ['Nuclear - % electricity', 'Fossil fuels - % electricity', 'Renewables - % electricity']

dados_filtrados_1 = filtrar_participacao(dados_1, fontes_energia_1)
dados_filtrados_2 = filtrar_participacao(dados_2, fontes_energia_2)

print("Gr√°fico para o primeiro dataset:")
gerar_grafico_comparacao(dados_filtrados_1, fontes_energia_1)

print("Gr√°fico para o segundo dataset:")
gerar_grafico_comparacao(dados_filtrados_2, fontes_energia_2)
```
